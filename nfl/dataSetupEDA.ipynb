```python
def offense_pbp(years):
    cols = ['posteam','posteam_type','yardline_100','dame_seconds_remaining','drive','down','ydstogo',
           'play_type','yards_gained','shotgun','qb_dropback','qb_kneel','qb_spike','qb_scramble',
           'pass_length','pass_location','air_yards','yards_after_catch','run_location','run_gap',
           'posteam_timeouts_remaining','defteam_timeouts_remaining','posteam_score','defteam_score',
           'score_differential','pass_attempt','penalty','time_to_throw','defenders_in_box',
           'offense_personnel','defense_personnel','number_of_pass_rushers','route','was_pressure','sack',
           'ngs_air_yards']
    df = nfl.import_pbp_data(years,cols)
    df['success'] = np.where((df.down==1 & df.yards_gained / df.ydstogo >=.4) |
                             (df.down==2 & df.yards_gained / df.ydstogo >=.6) |
                             (df.down==3 & df.yards_gained > = df.ydstogo),1,0)
    df['personal_13'] = np.where(df.offense_personnel.str.contains('1 RB, 3 TE'),1,0)
    df['personal_12'] = np.where(df.offense_personnel.str.contains('1 RB, 2 TE'),1,0)
    df['personal_11'] = np.where(df.offense_personnel.str.contains('1 RB, 1 TE'),1,0)
    df['personal_0x'] = np.where(df.offense_personnel.str.contains('0 RB, [1-3] TE'),1,0)
    df['personal_00'] = np.where(df.offense_personnel.str.contains('0 RB, 0 TE'),1,0)
    df['personal_10'] = np.where(df.offense_personnel.str.contains('1 RB, 0 TE'),1,0)
    df['personal_20'] = np.where(df.offense_personnel.str.contains('2 RB, 0 TE'),1,0)
    df['personal_21'] = np.where(df.offense_personnel.str.contains('2 RB, 1 TE'),1,0)
    df['personal_heavy'] = np.where(df.offense_personnel.str.contains('[2-3] RB, [2-3] TE'),1,0)
    df['passer_attmpt_depth'] = df.ngs_air_yards - df.air_yards
```
